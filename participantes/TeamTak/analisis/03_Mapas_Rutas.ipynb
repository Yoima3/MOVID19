# Mapa de personas sector salud que usan actualmente medios como patineta y bicicleta (Estratos 1-4)

```python
df_od = pd.read_csv('viajes_geo.csv', sep=';' )
```

```python

df_od_salud = df_od[(df_od['p7_id_actividad_economica']==17) & 
                    (df_od['p5_estrato'].isin([1,2,3,4]))  & 
                    (df_od['p17_Id_motivo_viaje'] != 'VOLVER CASA') & 
                    (df_od['modo_principal'].isin(['Bicicleta', 'Patineta']))]

default_map_loc=[4.638255, -74.085976]
map_salud_od = fl.Map(location=default_map_loc, zoom_start=12, tiles='cartodbpositron')
points_origen=[eval(i) for i in df_od_salud.coordenadas_origen.tolist()]
points_destino=[eval(i) for i in df_od_salud.coordenadas_destino.tolist()]

for i in range(0,len(points_origen)):
    fl.CircleMarker(location=points_origen[i], radius=2, color='green').add_to(map_salud_od)
    #fl.Marker(points_destino[i],icon=fl.Icon(color='gray')).add_to(map_salud_od)
    fl.CircleMarker(location=points_destino[i], radius=4, color='black').add_to(map_salud_od)
    fl.PolyLine([points_origen[i],points_destino[i]], color="red", weight=1.5, opacity=1).add_to(map_salud_od)
    
map_salud_od
```
